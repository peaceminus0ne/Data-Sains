```python
import csv
import pandas as pd
import numpy as py
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df=pd.read_csv('survei_media_sosial.csv')
df
```

```python
df.head(11)
```

```python
df.info()
```

```python
#Jenis kelamin pengguna media sosial
df['Jenis_Kelamin'].value_counts()
```

```python
ax=df['Jenis_Kelamin'].value_counts().plot.pie(figsize=(6,6), autopct='%.1f%%', fontsize=15)
plt.title('Jenis Kelamin Pengguna Media Sosial', fontsize=15)
plt.show()
```

```python
#Jenis Pekerjaan Pengguna Media Sosial
df['Pekerjaan'].value_counts()
```

```python
ax=df['Pekerjaan'].value_counts().plot.bar(figsize=(8,5), fontsize=12)
plt.title('Pekerjaan Pengguna Media Sosial', fontsize=25)
plt.show()
```

```python
#Media Sosial yang paling banyak digunakan oleh User
df['Sering'].value_counts()
```

```python
ax=df['Sering'].value_counts().plot.pie(figsize=(6,6), autopct='%.1f%%', fontsize=15)
plt.title('Media Sosial yang Sering Digunakan', fontsize=15)
plt.show()
```

```python
#Waktu Harian yang dihabiskan User untuk menggunakan Media Sosial
df['Waktu_Harian'].value_counts()
```

```python
ax=df['Waktu_Harian'].value_counts().plot.pie(figsize=(6,6), autopct='%.1f%%', fontsize=15)
plt.title('Waktu Harian yang Digunakan untuk bermain Sosial Media', fontsize=15)
plt.show()
```

```python
#Hasil survey dampak negatif penggunaan sosial media bagi user
negatif=df[['Mengganggu_Produktivitas', 'Fomo', 'Tanpa_Tujuan']]
negatif
```

```python
negatif=df[['Mengganggu_Produktivitas', 'Fomo', 'Tanpa_Tujuan']]
plt.hist(negatif, bins=7)
plt.title('Indikator Dampak Negatif Media Sosial bagi Pengguna', fontsize=15)
plt.show()
```

```python
#Hasil survey penyebab munculnya dampak negatif sosial media bagi user
diri=df[['Tidak_Bisa_Kontrol_Diri', 'Tidak_Sadar_Waktu']]
diri
```

```python
diri=df[['Tidak_Bisa_Kontrol_Diri', 'Tidak_Sadar_Waktu']]
plt.hist(diri, bins=7)
plt.title('Indikator Penyebab Munculnya Dampak Negatif Media Sosial bagi Pengguna', fontsize=15)
plt.show()
```

```python
#hasil survey usaha user untuk menghadadapi dampak negatif sosial media
usaha=df[['Usaha_Melepaskan_Diri', 'Pernah_Memakai_Aplikasi_Pengaturan_Waktu']]
usaha
```

```python
ax=df['Usaha_Melepaskan_Diri'].value_counts().plot.pie(figsize=(6,6), autopct='%.1f%%', fontsize=15)
plt.title('Usaha Melepaskan Diri dari Media Sosial', fontsize=15)
plt.show()
```

```python
ax=df['Pernah_Memakai_Aplikasi_Pengaturan_Waktu'].value_counts().plot.pie(figsize=(6,6), autopct='%.1f%%', fontsize=15)
plt.title('Usaha Melepaskan Diri dari Media Sosial dengan Bantuan Aplikasi Pengaturan Waktu', fontsize=15)
plt.show()
```
